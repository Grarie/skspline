```python
import matplotlib.pyplot as plt
```

```python
import numpy as np

from skspline import Spline
```

```python
x = np.linspace(0, 4, 1000)
y_ = 10*np.sin(x)

yerr = np.random.randn(len(y_))
y = y_ + yerr
```

```python
m = Spline(k=2)
m.fit(x, y)
```

```python
m.score(x,y)
```

```python
plt.scatter(x, y)

xx = np.linspace(0, 4, 5000)
yy = m.predict(xx)

plt.plot(xx,yy, color='k', linewidth=5)
```
